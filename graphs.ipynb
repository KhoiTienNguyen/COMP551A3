```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df= pd.read_csv('lr0_05.csv') 
df.columns = ["it", "relu", "leaky0.01","tanh","leaky0.25"]
dic ={"iteration" : df["it"] , "ReLU":df["relu"], "leaky ReLU γ = 0.01": df["leaky0.01"], "tanh": df["tanh"], "leaky ReLU γ = 0.25": df["leaky0.25"] }

df=pd.DataFrame(dic)

axes = plt.gca()

# pass the axes object to plot function
df.plot(kind='line', x='iteration', y='ReLU', ax=axes);
df.plot(kind='line', x='iteration', y='leaky ReLU γ = 0.01', ax=axes);
df.plot(kind='line', x='iteration', y='tanh', ax=axes);
df.plot(kind='line', x='iteration', y='leaky ReLU γ = 0.25', ax=axes);
plt.ylabel('Validation Accuracy')
plt.xlabel('Number of iterations')
plt.title('Validation accuracy with learning rate = 0.05')
```

```python
df= pd.read_csv('minibatch.csv') 
df.columns = ["it", "1", "2", "3", "4"]
dic ={"iteration" : df["it"] , "no decay, lr = 1.2":df["1"], "decay=0.995, lr = 1.2": df["2"], 
"no decay, lr = 1.5": df["3"], "ADAM, no decay, lr = 1": df["4"]  }

#,"Logistic with normalize & ADAM, no decay, lr = 1, mini-batch=16": df["4"]
df=pd.DataFrame(dic)
axes = plt.gca()
#axes.set_xlim([xmin, xmax])
#axes.set_ylim([0.625, 0.825])
# pass the axes object to plot function
df.plot(kind='line', x='iteration', y='no decay, lr = 1.2', ax=axes);
df.plot(kind='line', x='iteration', y='decay=0.995, lr = 1.2', ax=axes);
df.plot(kind='line', x='iteration', y='no decay, lr = 1.5', ax=axes);
df.plot(kind='line', x='iteration', y='ADAM, no decay, lr = 1', ax=axes);
plt.ylabel('Validation Accuracy')
plt.xlabel('Number of iterations')
plt.title('Validation accuracy with Logistic MLP over normalized images in mini batches of 16')

#iterations
#Logistic, no decay, lr = 1.2, mini-batch=16
#Logistic, decay=0.995, lr = 1.2, mini-batch=16
#Logistic, no decay, lr = 1.5, mini-batch=16
#Logistic with normalize & ADAM, no decay, lr = 1, mini-batch=16
```

```python
df= pd.read_csv('testtrain.csv') 
df.columns = ["it","train","test"]
dic ={"iteration" : df["it"] ,"Training Accuracy": df["train"] , "Test Accuracy": df["test"]}

df=pd.DataFrame(dic)
axes = plt.gca()
#axes.set_xlim([xmin, xmax])
#axes.set_ylim([0.625, 0.825])
# pass the axes object to plot function
df.plot(kind='line', x='iteration', y='Training Accuracy', ax=axes);
df.plot(kind='line', x='iteration', y='Test Accuracy', ax=axes);
plt.ylabel('Accuracy')
plt.xlabel('Number of iterations')
plt.title('Training and Test accuracy over Logistic MLP lr= 1.5 mini-batch =16')
```
